### Exercise 5: Subplot Layout

```python

import matplotlib.pyplot as plt
import numpy as np

try:
    fig, axes = plt.subplots(2, 2, figsize=(10, 8))

    x = np.linspace(0, 10, 100)
    y = x**2
    axes[0, 0].plot(x, y, color='b')
    axes[0, 0].set_title('Line Chart')

    categories = ['A', 'B', 'C', 'D']
    values = [5, 7, 3, 9]
    axes[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'purple'])
    axes[0, 1].set_title('Bar Chart')

    x_scatter = [1, 2, 3, 4, 5]
    y_scatter = [2, 4, 6, 8, 10]
    axes[1, 0].scatter(x_scatter, y_scatter, color='blue')
    for i in range(len(x_scatter)):
        axes[1, 0].annotate(f'({x_scatter[i]}, {y_scatter[i]})', (x_scatter[i], y_scatter[i]), textcoords="offset points", xytext=(5,5), ha='center')
    axes[1, 0].set_title('Scatter Plot')

    labels = ['Python', 'Java', 'C++', 'JavaScript']
    sizes = [40, 25, 20, 15]
    axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=['blue', 'red', 'green', 'purple'], startangle=140, wedgeprops={'edgecolor': 'black'})
    axes[1, 1].set_title('Pie Chart')

    plt.tight_layout()
    plt.show()

except ValueError as ve:
    print(f"Value Error: {ve}")
except TypeError as te:
    print(f"Type Error: {te}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
